```python
# Egy 5% fix, éves kuponú, 10 éves futamidejű, 100 Ft névértékű kötvény pénzáramlása


def bond_cash_flow(face_value, coupon_rate, years, coupon_frequency):
    cash_flows = []
    for i in range(1, years * coupon_frequency + 1):
        if i % coupon_frequency == 0:
            cash_flows.append(face_value * coupon_rate / coupon_frequency)
        else:
            cash_flows.append(0)
    cash_flows[-1] += face_value
    return cash_flows


import pandas as pd
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import numpy as np

face_value = 100
coupon_rate = 0.05
years = 10

cash_flows = bond_cash_flow(face_value, coupon_rate, years, 1)

df = pd.DataFrame(cash_flows, columns=['Cash Flow'])
df.index.name = 'Year'
display(df)

# plt.plot(cash_flows)
# plt.xlabel('Year')
# plt.ylabel('Cash Flow')
# plt.title('Bond Cash Flow')
# plt.show()

# # plot the cash flows on a bar chart

plt.bar(range(1, years + 1), cash_flows)
plt.xlabel('Year')
plt.ylabel('Cash Flow')
plt.title('Bond Cash Flow')
plt.show()
```

```python


def bond_cash_flow_stacked(face_value, coupon_rate, years, coupon_frequency):
    cash_flows1 = []
    cash_flows2 = []
    for i in range(1, years * coupon_frequency + 1):
        if i % coupon_frequency == 0:
            cash_flows1.append(face_value * coupon_rate / coupon_frequency)
            cash_flows2.append(0)
        else:
            cash_flows1.append(0)
            cash_flows2.append(0)
    cash_flows2[-1] = face_value
    return pd.DataFrame(np.array([cash_flows1, cash_flows2]), columns=np.arange(1, years + 1))



cash_flows = bond_cash_flow_stacked(face_value, coupon_rate, years, 1)
display(cash_flows)
cash_flows.T.plot.bar(stacked=True, legend=False)



plt.xlabel('Year')
plt.grid(axis='y',  linestyle='--')
plt.ylabel('Cash Flow')
plt.title('Bond Cash Flow')
plt.gcf().legend(loc='upper left', labels=['Kuponfizetés', 'Tőketörlesztés'])
plt.show()

```

```python
# Az egyes pénzáramlások jelenértéke 6% kamatláb mellett


def PV(r,n,fv):
    return fv/(1+r)**n


cash_flows = bond_cash_flow_stacked(face_value, coupon_rate, years, 1)
cash_flows = cash_flows.set_axis(["Kuponfizetés", "Tőketörlesztés"], axis=0)
display(cash_flows)

cash_flows.T.plot.bar(stacked=True, legend=True)

plt.xlabel('Year')
plt.grid(axis='y',  linestyle='--')
plt.ylabel('Cash Flow')
plt.title('Bond Cash Flow')
# plt.gcf().legend(loc='outside lower left', labels=['Kuponfizetés', 'Tőketörlesztés'])
plt.show()


cash_flows = cash_flows.set_axis([0, 1], axis=0)
cash_flows_PV = [[],[]]
for index, row in cash_flows.iterrows():
    for i in range(len(row)):
         cash_flows_PV[index].append( round( PV(0.06, i+1, row.iloc[i]), 4 ) )

df_PV = pd.DataFrame(np.array(cash_flows_PV), columns=np.arange(1, years + 1))
df_PV = df_PV.set_axis(["Kuponfizetés jelenértéke","Tőketörlesztés jelenértéke"], axis=0)
display(df_PV)


df_PV.T.plot.bar(stacked=True, legend=True)
plt.gca().set_ylim([0, 8])

plt.xlabel('Year')
plt.grid(axis='y',  linestyle='--')
plt.ylabel('Cash Flow')
plt.title('Bond Cash Flow PV')
# plt.gcf().legend(loc='outside lower left', labels=['Kuponfizetés jelenértéke', 'Tőketörlesztés jelenértéke'])
plt.show()

cash_flows_PV = [[],[]]
for index, row in cash_flows.iterrows():
    for i in range(len(row)):
         cash_flows_PV[index].append( round( PV(0.06, i+0.5, row.iloc[i]), 1 ) )

df_PV = pd.DataFrame(np.array(cash_flows_PV), columns=np.arange(1, years + 1))
df_PV = df_PV.set_axis(["Kuponfizetés jelenértéke","Tőketörlesztés jelenértéke"], axis=0)
display(df_PV)
display(Markdown('Megjegyzés: *Az MNB az oktatási füzeében az évek értékét a 0.5 évtől kezdve'
                 + ' helyetesítette a képletbe mindig 0,5 ével az egész érték előtt, és kerekítés egy tizedesre történt.*'))

df_PV.T.plot.bar(stacked=True, legend=True)
plt.gca().set_ylim([0, 8])

plt.xlabel('Year')
plt.grid(axis='y',  linestyle='--')
plt.ylabel('Cash Flow')
plt.title('Bond Cash Flow PV (MNB copy)')
# plt.gcf().legend(loc='outside lower left', labels=['Kuponfizetés jelenértéke', 'Tőketörlesztés jelenértéke'])
plt.show()
```
